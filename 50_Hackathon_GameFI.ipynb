```python
# prompt: Mystery Detective

import random

class MysteryDetective:

    def __init__(self):
        self.suspects = []
        self.clues = []
        self.evidence = []
        self.motive = None
        self.murder_weapon = None
        self.crime_scene = None
        self.victim = None

    def generate_suspects(self):
        suspects = ["John Smith", "Jane Doe", "Bob Jones", "Alice Brown", "Tom Green"]
        self.suspects = random.sample(suspects, 3)

    def generate_clues(self):
        clues = ["Bloody fingerprint", "Torn piece of clothing", "Mysterious note", "Broken alibi", "Witness testimony"]
        self.clues = random.sample(clues, 3)

    def generate_evidence(self):
        evidence = ["Murder weapon", "Victim's diary", "Crime scene photos", "Forensic report", "Witness statements"]
        self.evidence = random.sample(evidence, 3)

    def generate_motive(self):
        motives = ["Revenge", "Jealousy", "Greed", "Power", "Insanity"]
        self.motive = random.choice(motives)

    def generate_murder_weapon(self):
        murder_weapons = ["Gun", "Knife", "Blunt object", "Poison", "Strangulation"]
        self.murder_weapon = random.choice(murder_weapons)

    def generate_crime_scene(self):
        crime_scenes = ["Abandoned warehouse", "Dark alley", "Luxury apartment", "Deserted beach", "Haunted mansion"]
        self.crime_scene = random.choice(crime_scenes)

    def generate_victim(self):
        victims = ["Wealthy businessman", "Beautiful socialite", "Brilliant scientist", "Famous politician", "Innocent bystander"]
        self.victim = random.choice(victims)

    def solve_case(self):
        # Analyze suspects, clues, evidence, motive, murder weapon, crime scene, and victim to identify the killer.
        # Return the name of the killer.
        killer = random.choice(self.suspects)
        return killer

# Create a mystery detective object.
detective = MysteryDetective()

# Generate the case details.
detective.generate_suspects()
detective.generate_clues()
detective.generate_evidence()
detective.generate_motive()
detective.generate_murder_weapon()
detective.generate_crime_scene()
detective.generate_victim()

# Solve the case.
killer = detective.solve_case()

# Print the solution.
print("The killer is", killer)
```

```python
# prompt: Virtual Pet Simulator

import random

class VirtualPet:
    def __init__(self, name):
        self.name = name
        self.hunger = 50
        self.thirst = 50
        self.energy = 50
        self.happiness = 50

    def feed(self):
        self.hunger -= 10
        self.thirst += 5
        self.energy += 5
        self.happiness += 5

    def water(self):
        self.thirst -= 10
        self.hunger += 5
        self.energy += 5
        self.happiness += 5

    def play(self):
        self.energy -= 10
        self.hunger += 5
        self.thirst += 5
        self.happiness += 10

    def sleep(self):
        self.energy += 10
        self.hunger += 5
        self.thirst += 5
        self.happiness += 5

    def status(self):
        print(f"{self.name}'s hunger: {self.hunger}")
        print(f"{self.name}'s thirst: {self.thirst}")
        print(f"{self.name}'s energy: {self.energy}")
        print(f"{self.name}'s happiness: {self.happiness}")

    def check_alive(self):
        if self.hunger <= 0 or self.thirst <= 0 or self.energy <= 0 or self.happiness <= 0:
            return False
        else:
            return True

# Create a virtual pet.
pet = VirtualPet("Fluffy")

# Main loop.
while pet.check_alive():
    # Print the pet's status.
    pet.status()

    # Ask the user what they want to do.
    action = input("What do you want to do? (feed, water, play, sleep) ")

    # Perform the action.
    if action == "feed":
        pet.feed()
    elif action == "water":
        pet.water()
    elif action == "play":
        pet.play()
    elif action == "sleep":
        pet.sleep()
    else:
        print("Invalid action.")

# Print a message when the pet dies.
if not pet.check_alive():
    print(f"{pet.name} has died.")
```

```python
# prompt: Alien Invasion

import random

# Create a list of possible alien names.
alien_names = ["Zorblax", "Glorblax", "Flarblax", "Blorblax", "Klorblax"]

# Create a list of possible alien weaknesses.
alien_weaknesses = ["Water", "Fire", "Ice", "Electricity", "Sonic"]

# Create a function to generate a random alien.
def generate_alien():
  name = random.choice(alien_names)
  weakness = random.choice(alien_weaknesses)
  health = random.randint(10, 20)
  return {"name": name, "weakness": weakness, "health": health}

# Create a list of aliens.
aliens = []
for i in range(5):
  aliens.append(generate_alien())

# Create a loop that will continue until all aliens are defeated.
while len(aliens) > 0:

  # Print the list of aliens.
  for i, alien in enumerate(aliens):
    print(f"{i+1}. {alien['name']} (Health: {alien['health']})")

  # Ask the player which alien they want to attack.
  choice = int(input("Which alien do you want to attack? (Enter a number) "))

  # Get the chosen alien.
  chosen_alien = aliens[choice-1]

  # Ask the player what weapon they want to use.
  weapon = input("What weapon do you want to use? (Water, Fire, Ice, Electricity, Sonic) ")

  # Check if the chosen weapon is effective against the alien.
  if weapon == chosen_alien["weakness"]:
    damage = 10
  else:
    damage = 5

  # Decrease the alien's health by the damage.
  chosen_alien["health"] -= damage

  # Print a message indicating the damage dealt.
  print(f"You dealt {damage} damage to {chosen_alien['name']}.")

  # If the alien's health reaches 0, remove it from the list.
  if chosen_alien["health"] <= 0:
    aliens.remove(chosen_alien)
    print(f"{chosen_alien['name']} has been defeated.")

# Print a message when all aliens are defeated.
if len(aliens) == 0:
  print("You have defeated all the aliens!")
```

```python
# prompt: Underwater Exploration

import random

class UnderwaterExplorer:
    def __init__(self, name):
        self.name = name
        self.oxygen_level = 100
        self.depth = 0

    def explore(self):
        # Explore a random location underwater.
        location = random.choice(["Coral Reef", "Sunken Ship", "Underwater Cave"])
        print(f"{self.name} is exploring a {location}.")

        # Decrease oxygen level.
        self.oxygen_level -= 10

        # Increase depth.
        self.depth += 10

    def resurface(self):
        # Resurface and refill oxygen level.
        self.oxygen_level = 100
        self.depth = 0
        print(f"{self.name} has resurfaced.")

    def check_oxygen(self):
        if self.oxygen_level <= 0:
            print(f"{self.name} has run out of oxygen and drowned.")
            return False
        else:
            return True

# Create an underwater explorer.
explorer = UnderwaterExplorer("Jacques Cousteau")

# Explore underwater until the explorer runs out of oxygen.
while explorer.check_oxygen():
    explorer.explore()

    # Check if the explorer needs to resurface.
    if explorer.oxygen_level <= 30:
        explorer.resurface()

# Print a message when the explorer runs out of oxygen.
print(f"{explorer.name} has died.")
```

```python
# prompt: Magic Academy

class MagicStudent:
    def __init__(self, name, health, mana, spellbook):
        self.name = name
        self.health = health
        self.mana = mana
        self.spellbook = spellbook

    def cast_spell(self, spell, opponent):
        # Check if the student has enough mana to cast the spell.
        if self.mana >= spell.mana_cost:
            # Decrease the student's mana by the spell's mana cost.
            self.mana -= spell.mana_cost

            # Calculate the damage dealt by the spell.
            damage = spell.damage + random.randint(1, 6)

            # Decrease the opponent's health by the damage dealt.
            opponent.health -= damage

            # Print a message indicating the damage dealt.
            print(f"{self.name} cast {spell.name} on {opponent.name} for {damage} damage.")
        else:
            print(f"{self.name} does not have enough mana to cast {spell.name}.")

    def is_alive(self):
        return self.health > 0

class Spell:
    def __init__(self, name, damage, mana_cost):
        self.name = name
        self.damage = damage
        self.mana_cost = mana_cost

# Create a spellbook.
spellbook = [
    Spell("Fireball", 10, 5),
    Spell("Ice Shard", 5, 3),
    Spell("Lightning Bolt", 15, 8),
    Spell("Heal", -5, 5)
]

# Create two magic students.
student1 = MagicStudent("Harry Potter", 100, 100, spellbook)
student2 = MagicStudent("Hermione Granger", 80, 80, spellbook)

# Start the duel.
while student1.is_alive() and student2.is_alive():
    # Choose a random student to cast a spell.
    caster = random.choice([student1, student2])

    # Choose a random student to defend.
    defender = random.choice([student1, student2])

    # Choose a random spell from the caster's spellbook.
    spell = random.choice(caster.spellbook)

    # Cast the spell.
    caster.cast_spell(spell, defender)

    # Check if the duel is over.
    if not student1.is_alive():
        print(f"{student2.name} wins the duel!")
    elif not student2.is_alive():
        print(f"{student1.name} wins the duel!")
```

```python
# prompt: Pirate Treasure Hunt

import random

# Create a list of possible treasure locations.
locations = ["Isla Nublar", "Skull Island", "Treasure Island", "Tortuga", "Cayo Coco"]

# Create a list of possible treasures.
treasures = ["Gold Doubloons", "Silver Bars", "Jewels", "Ancient Artifacts", "Map to the Fountain of Youth"]

# Create a function to generate a random treasure hunt.
def generate_treasure_hunt():
  # Choose a random location.
  location = random.choice(locations)

  # Choose a random treasure.
  treasure = random.choice(treasures)

  # Generate a random number of clues.
  num_clues = random.randint(3, 5)

  # Create a list of clues.
  clues = []
  for i in range(num_clues):
    clue = f"Clue {i+1}: {random.choice(['Follow the X marks the spot', 'Look for the buried treasure chest', 'Ask the locals for directions', 'Search the ancient ruins', 'Follow the rainbow'])}"
    clues.append(clue)

  # Return the treasure hunt.
  return {"location": location, "treasure": treasure, "clues": clues}

# Create a treasure hunt.
treasure_hunt = generate_treasure_hunt()

# Print the treasure hunt information.
print(f"Treasure Hunt Location: {treasure_hunt['location']}")
print(f"Treasure: {treasure_hunt['treasure']}")
print("Clues:")
for clue in treasure_hunt["clues"]:
  print(clue)

# Ask the player if they want to start the treasure hunt.
start = input("Do you want to start the treasure hunt? (y/n) ")

# If the player wants to start the treasure hunt, start the game loop.
if start == "y":
  # Loop until the player finds the treasure.
  while True:
    # Print the current clue.
    clue = treasure_hunt["clues"].pop(0)
    print(clue)

    # Ask the player if they want to follow the clue.
    follow = input("Do you want to follow the clue? (y/n) ")

    # If the player wants to follow the clue, check if they found the treasure.
    if follow == "y":
      # Generate a random number to determine if the player found the treasure.
      found_treasure = random.randint(1, 2)

      # If the player found the treasure, print a message and break out of the loop.
      if found_treasure == 1:
        print(f"You found the treasure! It is a {treasure_hunt['treasure']}!")
        break

      # If the player did not find the treasure, print a message and continue the loop.
      else:
        print("You did not find the treasure. Keep searching!")

    # If the player does not want to follow the clue, print a message and continue the loop.
    else:
      print("You did not follow the clue. Keep searching!")

# If the player does not want to start the treasure hunt, print a message.
else:
  print("You did not start the treasure hunt.")
```

```python
# prompt: Space Colony Builder

import random

# Define the resources available to the colony.
resources = {
    "water": 100,
    "food": 100,
    "oxygen": 100,
    "energy": 100
}

# Define the population of the colony.
population = 100

# Define the number of turns the colony has survived.
turns = 0

# Create a list of possible events that can occur.
events = [
    "A meteor shower damages the colony's oxygen supply.",
    "A disease outbreak reduces the population of the colony.",
    "A new source of food is discovered, increasing the colony's food supply.",
    "A new energy source is discovered, increasing the colony's energy supply.",
    "A new water source is discovered, increasing the colony's water supply."
]

# Start the game loop.
while True:
    # Print the current status of the colony.
    print(f"Turn: {turns}")
    print(f"Population: {population}")
    print(f"Resources:")
    for resource, amount in resources.items():
        print(f"  {resource}: {amount}")

    # Generate a random event.
    event = random.choice(events)

    # Handle the event.
    if event == "A meteor shower damages the colony's oxygen supply.":
        resources["oxygen"] -= random.randint(10, 20)
    elif event == "A disease outbreak reduces the population of the colony.":
        population -= random.randint(10, 20)
    elif event == "A new source of food is discovered, increasing the colony's food supply.":
        resources["food"] += random.randint(10, 20)
    elif event == "A new energy source is discovered, increasing the colony's energy supply.":
        resources["energy"] += random.randint(10, 20)
    elif event == "A new water source is discovered, increasing the colony's water supply.":
        resources["water"] += random.randint(10, 20)

    # Check if the colony has survived.
    if population <= 0 or any(amount <= 0 for amount in resources.values()):
        print("The colony has failed.")
        break

    # Increase the turn counter.
    turns += 1
```

```python
# prompt: City Builder

class City:
    def __init__(self, name):
        self.name = name
        self.population = 100
        self.resources = {"food": 100, "wood": 100, "stone": 100}
        self.buildings = []

    def grow_population(self):
        self.population += 10

    def gather_resources(self):
        for resource in self.resources:
            self.resources[resource] += 10

    def build_house(self):
        if self.resources["wood"] >= 10 and self.resources["stone"] >= 10:
            self.resources["wood"] -= 10
            self.resources["stone"] -= 10
            self.buildings.append("house")

    def build_farm(self):
        if self.resources["wood"] >= 20 and self.resources["stone"] >= 20:
            self.resources["wood"] -= 20
            self.resources["stone"] -= 20
            self.buildings.append("farm")

    def build_mine(self):
        if self.resources["wood"] >= 30 and self.resources["stone"] >= 30:
            self.resources["wood"] -= 30
            self.resources["stone"] -= 30
            self.buildings.append("mine")

    def print_status(self):
        print(f"City: {self.name}")
        print(f"Population: {self.population}")
        print(f"Resources:")
        for resource, amount in self.resources.items():
            print(f"  {resource}: {amount}")
        print(f"Buildings: {', '.join(self.buildings)}")

# Create a city.
city = City("New York")

# Run the game loop.
while True:
    # Print the city's status.
    city.print_status()

    # Ask the player what they want to do.
    action = input("What do you want to do? (grow population/gather resources/build house/build farm/build mine/quit) ")

    # Handle the player's action.
    if action == "grow population":
        city.grow_population()
    elif action == "gather resources":
        city.gather_resources()
    elif action == "build house":
        city.build_house()
    elif action == "build farm":
        city.build_farm()
    elif action == "build mine":
        city.build_mine()
    elif action == "quit":
        break

# Print a message indicating that the game has ended.
print("Game over.")
```

```python
# prompt: Monster Tamer

import random

class Monster:
    def __init__(self, name, health, attack, defense):
        self.name = name
        self.health = health
        self.attack = attack
        self.defense = defense

    def attack(self, opponent):
        damage = self.attack - opponent.defense
        if damage < 0:
            damage = 0
        opponent.health -= damage
        print(f"{self.name} dealt {damage} damage to {opponent.name}.")

    def is_alive(self):
        return self.health > 0

class Trainer:
    def __init__(self, name):
        self.name = name
        self.monsters = []

    def add_monster(self, monster):
        self.monsters.append(monster)

    def attack(self, opponent):
        # Choose a random monster to attack.
        attacker = random.choice(self.monsters)

        # Attack the opponent.
        attacker.attack(opponent)

        # Check if the opponent is dead.
        if not opponent.is_alive():
            print(f"{opponent.name} has been defeated.")

# Create a list of monster names.
monster_names = ["Charmander", "Squirtle", "Bulbasaur", "Pikachu", "Eevee"]

# Create a list of monster attributes.
monster_attributes = {
    "health": [100, 80, 60, 40, 20],
    "attack": [10, 8, 6, 4, 2],
    "defense": [5, 4, 3, 2, 1]
}

# Create a list of trainers.
trainers = []
for i in range(2):
    name = input(f"Enter the name of trainer {i+1}: ")
    trainer = Trainer(name)

    # Create a list of monsters for the trainer.
    for j in range(3):
        name = random.choice(monster_names)
        health = random.choice(monster_attributes["health"])
        attack = random.choice(monster_attributes["attack"])
        defense = random.choice(monster_attributes["defense"])
        monster = Monster(name, health, attack, defense)
        trainer.add_monster(monster)

    trainers.append(trainer)

# Start the battle.
while True:
    # Choose a random trainer to attack.
    attacker = random.choice(trainers)

    # Choose a random opponent trainer.
    opponent = random.choice(trainers)

    # Check if the attacker and opponent are the same trainer.
    if attacker == opponent:
        continue

    # Attack the opponent.
    attacker.attack(opponent)

    # Check if the opponent has any monsters left.
    if not any(monster.is_alive() for monster in opponent.monsters):
        print(f"{opponent.name} has no monsters left and has been defeated.")
        trainers.remove(opponent)

    # Check if there is only one trainer left.
    if len(trainers) == 1:
        print(f"{trainers[0].name} is the winner!")
        break
```

```python
# prompt: Fantasy Card Game

class Card:
    def __init__(self, name, cost, power, toughness):
        self.name = name
        self.cost = cost
        self.power = power
        self.toughness = toughness

class Player:
    def __init__(self, name, health):
        self.name = name
        self.health = health
        self.hand = []
        self.mana = 0

    def draw_card(self, card):
        self.hand.append(card)

    def play_card(self, card):
        if card.cost <= self.mana:
            self.mana -= card.cost
            self.hand.remove(card)
        else:
            print("Not enough mana to play card.")

    def attack(self, opponent, card):
        opponent.health -= card.power

class Game:
    def __init__(self):
        self.players = []
        self.turn = 0

    def add_player(self, player):
        self.players.append(player)

    def start_game(self):
        # Deal cards to players.
        for player in self.players:
            for i in range(7):
                player.draw_card(Card("Forest", 1, 0, 0))

        # Set starting mana.
        for player in self.players:
            player.mana = 1

    def next_turn(self):
        self.turn = (self.turn + 1) % len(self.players)

    def play_turn(self):
        player = self.players[self.turn]

        # Draw a card.
        player.draw_card(Card("Forest", 1, 0, 0))

        # Increase mana.
        player.mana += 1

        # Play a card.
        card = player.hand[0]
        player.play_card(card)

        # Attack opponent.
        opponent = self.players[(self.turn + 1) % len(self.players)]
        player.attack(opponent, card)

        # Check if game is over.
        if opponent.health <= 0:
            print(f"{player.name} wins!")
            return

        # End turn.
        self.next_turn()

# Create a game.
game = Game()

# Create players.
player1 = Player("Player 1", 20)
player2 = Player("Player 2", 20)

# Add players to game.
game.add_player(player1)
game.add_player(player2)

# Start the game.
game.start_game()

# Play the game.
while True:
    game.play_turn()
```

```python
# prompt: Cyber Heist

import random

class Person:
    def __init__(self, name):
        self.name = name
        self.skills = {}

class Hacker(Person):
    def __init__(self, name):
        super().__init__(name)
        self.skills["hacking"] = random.randint(1, 10)

class SecurityGuard(Person):
    def __init__(self, name):
        super().__init__(name)
        self.skills["security"] = random.randint(1, 10)

class CyberHeist:
    def __init__(self, hackers, guards):
        self.hackers = hackers
        self.guards = guards
        self.progress = 0

    def attempt_hack(self):
        # Choose a random hacker.
        hacker = random.choice(self.hackers)

        # Choose a random guard.
        guard = random.choice(self.guards)

        # Roll a die to determine the outcome of the hack attempt.
        roll = random.randint(1, 20)

        # If the hacker's skill is higher than the guard's skill, the hack attempt succeeds.
        if hacker.skills["hacking"] > guard.skills["security"]:
            self.progress += roll
            print(f"{hacker.name} successfully hacked the system and made progress.")
        else:
            print(f"{hacker.name} failed to hack the system.")

    def attempt_guard(self):
        # Choose a random guard.
        guard = random.choice(self.guards)

        # Choose a random hacker.
        hacker = random.choice(self.hackers)

        # Roll a die to determine the outcome of the guard's attempt.
        roll = random.randint(1, 20)

        # If the guard's skill is higher than the hacker's skill, the guard catches the hacker.
        if guard.skills["security"] > hacker.skills["hacking"]:
            self.hackers.remove(hacker)
            print(f"{guard.name} caught {hacker.name} and removed them from the heist.")
        else:
            print(f"{guard.name} failed to catch {hacker.name}.")

    def is_successful(self):
        return self.progress >= 100

    def play_turn(self):
        # Each hacker attempts to hack.
        for hacker in self.hackers:
            self.attempt_hack()

        # Each guard attempts to catch a hacker.
        for guard in self.guards:
            self.attempt_guard()

        # Check if the heist is successful.
        if self.is_successful():
            print("The heist was successful!")
            return True

        # Check if the heist has failed.
        if len(self.hackers) == 0:
            print("The heist failed.")
            return False

        # Continue to the next turn.
        return None

# Create some hackers and guards.
hackers = [Hacker("Hacker 1"), Hacker("Hacker 2"), Hacker("Hacker 3")]
guards = [SecurityGuard("Guard 1"), SecurityGuard("Guard 2"), SecurityGuard("Guard 3")]

# Create a cyber heist.
heist = CyberHeist(hackers, guards)

# Play the heist until it is successful or fails.
while True:
    if heist.play_turn():
        break
```

```python
# prompt: Magic Duel

class Spell:
    def __init__(self, name, damage, mana_cost):
        self.name = name
        self.damage = damage
        self.mana_cost = mana_cost

class Wizard:
    def __init__(self, name, health, mana):
        self.name = name
        self.health = health
        self.mana = mana
        self.spells = [Spell("Fireball", 10, 5), Spell("Ice Shard", 5, 3), Spell("Lightning Bolt", 15, 8)]

    def cast_spell(self, spell, opponent):
        if spell.mana_cost > self.mana:
            print(f"{self.name} does not have enough mana to cast {spell.name}.")
            return
        self.mana -= spell.mana_cost
        opponent.health -= spell.damage
        print(f"{self.name} casts {spell.name} on {opponent.name} for {spell.damage} damage.")

    def is_alive(self):
        return self.health > 0

def duel(wizard1, wizard2):
    while wizard1.is_alive() and wizard2.is_alive():
        # Wizard 1 casts a spell on wizard 2.
        wizard1.cast_spell(random.choice(wizard1.spells), wizard2)

        # Check if wizard 2 is dead.
        if not wizard2.is_alive():
            break

        # Wizard 2 casts a spell on wizard 1.
        wizard2.cast_spell(random.choice(wizard2.spells), wizard1)

    # Print the results of the duel.
    if wizard1.is_alive():
        print(f"{wizard1.name} wins!")
    else:
        print(f"{wizard2.name} wins!")

# Create two wizards.
wizard1 = Wizard("Merlin", 100, 50)
wizard2 = Wizard("Gandalf", 120, 60)

# Start the duel.
duel(wizard1, wizard2)
```

```python
# prompt: Space Trading

import random

class TradeGood:
    def __init__(self, name, buy_price, sell_price):
        self.name = name
        self.buy_price = buy_price
        self.sell_price = sell_price

class Planet:
    def __init__(self, name, goods):
        self.name = name
        self.goods = goods

class Ship:
    def __init__(self, name, cargo_capacity):
        self.name = name
        self.cargo_capacity = cargo_capacity
        self.cargo = {}

class Trader:
    def __init__(self, name, ship):
        self.name = name
        self.ship = ship
        self.credits = 1000

    def buy_goods(self, planet, good, quantity):
        if good not in planet.goods:
            print(f"{good} is not available on {planet.name}.")
            return
        if quantity > planet.goods[good]:
            print(f"There is not enough {good} available on {planet.name}.")
            return
        if self.credits < good.buy_price * quantity:
            print(f"You do not have enough credits to buy {quantity} {good.name}.")
            return
        self.ship.cargo[good] = self.ship.cargo.get(good, 0) + quantity
        self.credits -= good.buy_price * quantity
        planet.goods[good] -= quantity

    def sell_goods(self, planet, good, quantity):
        if good not in self.ship.cargo:
            print(f"You do not have any {good.name} to sell.")
            return
        if quantity > self.ship.cargo[good]:
            print(f"You do not have enough {good.name} to sell.")
            return
        self.credits += good.sell_price * quantity
        self.ship.cargo[good] -= quantity
        planet.goods[good] = planet.goods.get(good, 0) + quantity

    def travel(self, planet):
        self.ship.location = planet

# Create some trade goods.
wheat = TradeGood("Wheat", 10, 15)
ore = TradeGood("Ore", 20, 30)
fuel = TradeGood("Fuel", 5, 10)

# Create some planets.
earth = Planet("Earth", {wheat: 100, ore: 50, fuel: 25})
mars = Planet("Mars", {wheat: 50, ore: 100, fuel: 75})

# Create a ship.
ship = Ship("Millennium Falcon", 100)

# Create a trader.
trader = Trader("Han Solo", ship)

# Set the ship's location to Earth.
ship.location = earth

# Start the trading loop.
while True:
    # Print the trader's current location and credits.
    print(f"You are currently on {ship.location.name}. You have {trader.credits} credits.")

    # Print the available goods on the current planet.
    print(f"Available goods:")
    for good, quantity in ship.location.goods.items():
        print(f"- {good.name}: {quantity}")

    # Ask the trader what they want to do.
    action = input("What do you want to do? (buy, sell, travel, quit) ")

    # Handle the trader's action.
    if action == "buy":
        # Ask the trader what good they want to buy.
        good_name = input("What good do you want to buy? ")

        # Find the good in the planet's goods.
        good = None
        for g in ship.location.goods:
            if g.name == good_name:
                good = g
                break

        # If the good is not available, print an error message.
        if good is None:
            print(f"{good_name} is not available on {ship.location.name}.")
            continue

        # Ask the trader how much of the good they want to buy.
        quantity = int(input(f"How much {good.name} do you want to buy? "))

        # Buy the good.
        trader.buy_goods(ship.location, good, quantity)
    elif action == "sell":
        # Ask the trader what good they want to sell.
        good_name = input("What good do you want to sell? ")

        # Find the good in the ship's cargo.
        good = None
        for g in ship.cargo:
            if g.name == good_name:
                good = g
                break

        # If the good is not available, print an error message.
        if good is None:
            print(f"You do not have any {good_name} to sell.")
            continue

        # Ask the trader how much of the good they want to sell.
        quantity = int(input(f"How much {good.name} do you want to sell? "))

        # Sell the good.
        trader.sell_goods(ship.location, good, quantity)
    elif action == "travel":
        # Ask the trader what planet they want to travel to.
        planet_name = input("What planet do you want to travel to? ")

        # Find the planet.
        planet = None
        for p in [earth, mars]:
            if p.name == planet_name:
                planet = p
                break

        # If the planet is not found, print an error message.
        if planet is None:
            print(f"{planet_name} is not a valid planet.")
            continue

        # Travel to the planet.
        trader.travel(planet)
    elif action == "quit":
        # Exit the trading loop.
        break
    else:
        # Print an error message.
        print("Invalid action.")

# Print a message indicating that the trader has quit.
print("You have quit the game.")
```

```python
# prompt: Virtual Farming

class Crop:
    def __init__(self, name, growth_rate, yield):
        self.name = name
        self.growth_rate = growth_rate
        self.yield = yield
        self.age = 0

    def grow(self):
        self.age += 1

    def is_mature(self):
        return self.age >= self.growth_rate

    def harvest(self):
        if self.is_mature():
            return self.yield
        else:
            return 0

class Farm:
    def __init__(self, name):
        self.name = name
        self.crops = []

    def add_crop(self, crop):
        self.crops.append(crop)

    def grow_crops(self):
        for crop in self.crops:
            crop.grow()

    def harvest_crops(self):
        yield_total = 0
        for crop in self.crops:
            yield_total += crop.harvest()
        return yield_total

class Farmer:
    def __init__(self, name):
        self.name = name
        self.farm = Farm(name + "'s Farm")

    def add_crop(self, crop):
        self.farm.add_crop(crop)

    def grow_crops(self):
        self.farm.grow_crops()

    def harvest_crops(self):
        return self.farm.harvest_crops()

# Create some crops.
wheat = Crop("Wheat", 3, 10)
corn = Crop("Corn", 4, 15)
tomatoes = Crop("Tomatoes", 2, 5)

# Create a farmer.
farmer = Farmer("John")

# Add crops to the farmer's farm.
farmer.add_crop(wheat)
farmer.add_crop(corn)
farmer.add_crop(tomatoes)

# Grow the crops for a few days.
for i in range(5):
    farmer.grow_crops()

# Harvest the crops.
yield_total = farmer.harvest_crops()

# Print the total yield.
print(f"Total yield: {yield_total}")
```

```python
# prompt: Time Management

import datetime

def time_management(tasks, durations):
  """
  This function takes two lists, tasks and durations, and returns a list of tuples containing the start and end times for each task.

  Args:
    tasks: A list of strings representing the names of the tasks.
    durations: A list of integers representing the durations of the tasks in minutes.

  Returns:
    A list of tuples containing the start and end times for each task.
  """

  start_times = []
  end_times = []

  current_time = datetime.datetime.now()

  for duration in durations:
    start_times.append(current_time)
    current_time += datetime.timedelta(minutes=duration)
    end_times.append(current_time)

  return list(zip(start_times, end_times))

tasks = ["Task 1", "Task 2", "Task 3"]
durations = [30, 45, 60]

schedule = time_management(tasks, durations)

for task, (start_time, end_time) in zip(tasks, schedule):
  print(f"{task}: {start_time.strftime('%H:%M')} - {end_time.strftime('%H:%M')}")
```

```python
# prompt: Fantasy Battle

import random

class Character:
    def __init__(self, name, health, attack_power):
        self.name = name
        self.health = health
        self.attack_power = attack_power

    def attack(self, opponent):
        damage = random.randint(0, self.attack_power)
        opponent.health -= damage
        print(f"{self.name} attacks {opponent.name} for {damage} damage.")

    def is_alive(self):
        return self.health > 0

def fantasy_battle(character1, character2):
    while character1.is_alive() and character2.is_alive():
        # Character 1 attacks character 2.
        character1.attack(character2)

        # Check if character 2 is dead.
        if not character2.is_alive():
            break

        # Character 2 attacks character 1.
        character2.attack(character1)

    # Print the results of the battle.
    if character1.is_alive():
        print(f"{character1.name} wins!")
    else:
        print(f"{character2.name} wins!")

# Create two characters.
character1 = Character("Warrior", 100, 10)
character2 = Character("Wizard", 80, 15)

# Start the battle.
fantasy_battle(character1, character2)
```

```python
# prompt: Alien Research Lab

class Alien:
    def __init__(self, species, home_planet, abilities):
        self.species = species
        self.home_planet = home_planet
        self.abilities = abilities

class ResearchLab:
    def __init__(self, name):
        self.name = name
        self.aliens = []

    def add_alien(self, alien):
        self.aliens.append(alien)

    def study_alien(self, alien):
        print(f"Studying {alien.species} from {alien.home_planet}.")
        for ability in alien.abilities:
            print(f"- {ability}")

# Create some aliens.
alien1 = Alien("Vulcan", "Vulcan", ["telepathy", "logic"])
alien2 = Alien("Klingon", "Qo'noS", ["strength", "endurance"])
alien3 = Alien("Romulan", "Romulus", ["cunning", "deception"])

# Create a research lab.
lab = ResearchLab("Starfleet Research Lab")

# Add aliens to the research lab.
lab.add_alien(alien1)
lab.add_alien(alien2)
lab.add_alien(alien3)

# Study an alien.
lab.study_alien(alien1)
```

```python
# prompt: Magic World Builder

class Spell:
    def __init__(self, name, power, element):
        self.name = name
        self.power = power
        self.element = element

class Wizard:
    def __init__(self, name, health, spells):
        self.name = name
        self.health = health
        self.spells = spells

    def cast_spell(self, spell, target):
        damage = spell.power
        if spell.element == "fire" and target.element == "water":
            damage *= 2
        elif spell.element == "water" and target.element == "fire":
            damage *= 0.5
        target.health -= damage
        print(f"{self.name} casts {spell.name} on {target.name} for {damage} damage.")

class Creature:
    def __init__(self, name, health, element):
        self.name = name
        self.health = health
        self.element = element

def magic_battle(wizard1, wizard2):
    while wizard1.health > 0 and wizard2.health > 0:
        # Wizard 1 attacks wizard 2.
        spell = random.choice(wizard1.spells)
        wizard1.cast_spell(spell, wizard2)

        # Check if wizard 2 is dead.
        if wizard2.health <= 0:
            break

        # Wizard 2 attacks wizard 1.
        spell = random.choice(wizard2.spells)
        wizard2.cast_spell(spell, wizard1)

    # Print the results of the battle.
    if wizard1.health > 0:
        print(f"{wizard1.name} wins!")
    else:
        print(f"{wizard2.name} wins!")

# Create some spells.
fireball = Spell("Fireball", 10, "fire")
ice_bolt = Spell("Ice Bolt", 8, "water")
lightning_strike = Spell("Lightning Strike", 12, "lightning")

# Create some wizards.
wizard1 = Wizard("Gandalf", 100, [fireball, ice_bolt])
wizard2 = Wizard("Saruman", 90, [lightning_strike, ice_bolt])

# Create some creatures.
creature1 = Creature("Dragon", 120, "fire")
creature2 = Creature("Troll", 80, "earth")

# Start a magic battle.
magic_battle(wizard1, wizard2)

```

```python
# prompt: Space Rescue Mission

import random

class Spacecraft:
    def __init__(self, name, speed, fuel_capacity):
        self.name = name
        self.speed = speed
        self.fuel_capacity = fuel_capacity
        self.current_fuel = fuel_capacity

    def travel(self, distance):
        fuel_used = distance / self.speed
        if fuel_used > self.current_fuel:
            raise ValueError("Not enough fuel to travel that distance.")
        self.current_fuel -= fuel_used

class Astronaut:
    def __init__(self, name, health):
        self.name = name
        self.health = health

class SpaceStation:
    def __init__(self, name, distance_from_earth):
        self.name = name
        self.distance_from_earth = distance_from_earth
        self.astronauts = []

    def add_astronaut(self, astronaut):
        self.astronauts.append(astronaut)

class SpaceRescueMission:
    def __init__(self, spacecraft, astronauts, space_station):
        self.spacecraft = spacecraft
        self.astronauts = astronauts
        self.space_station = space_station

    def launch(self):
        print(f"Spacecraft {self.spacecraft.name} launched to space station {self.space_station.name}.")
        self.spacecraft.travel(self.space_station.distance_from_earth)
        print(f"Spacecraft {self.spacecraft.name} arrived at space station {self.space_station.name}.")

    def rescue_astronauts(self):
        for astronaut in self.space_station.astronauts:
            print(f"Rescuing astronaut {astronaut.name} from space station {self.space_station.name}.")
            astronaut.health = 100

    def return_to_earth(self):
        print(f"Spacecraft {self.spacecraft.name} returning to Earth.")
        self.spacecraft.travel(self.space_station.distance_from_earth)
        print(f"Spacecraft {self.spacecraft.name} landed on Earth.")

# Create a spacecraft.
spacecraft = Spacecraft("Discovery", 17000, 72)

# Create some astronauts.
astronauts = [
    Astronaut("John Smith", 80),
    Astronaut("Jane Doe", 75),
    Astronaut("Michael Jones", 90),
]

# Create a space station.
space_station = SpaceStation("ISS", 400000)

# Add astronauts to the space station.
for astronaut in astronauts:
    space_station.add_astronaut(astronaut)

# Create a space rescue mission.
mission = SpaceRescueMission(spacecraft, astronauts, space_station)

# Launch the mission.
mission.launch()

# Rescue the astronauts.
mission.rescue_astronauts()

# Return to Earth.
mission.return_to_earth()
```
